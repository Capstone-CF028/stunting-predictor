```python
#menghapus data duplikat 
df = df.drop_duplicates()
duplicate = df.duplicated().sum()
print(f'Jumlah data duplikat : {duplicate}')
```

```python
#Mengelompokkan kolom numerik dan kategori
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
print(numerical_cols)
print(categorical_cols)
```

```python
# Normalisasi Fitur Numerik dengan MinMaxScaler
scaler = MinMaxScaler()
df[numerical_cols] = scaler.fit_transform(df[numerical_cols])
```

```python
# Label Encoding untuk Fitur Kategorikal
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
df['Jenis Kelamin'] = le.fit_transform(df['Jenis Kelamin'])
```

```python
print(df['Wasting'].value_counts())
```

```python
# Pisahkan fitur (X) dan label (y)
X = df.drop(columns=["Stunting", 'Wasting'])  # semua kolom kecuali target
y = df["Wasting"]                 # kolom target
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

# Asumsikan X adalah fitur dan y adalah label (kategori stunting)
print("Distribusi sebelum SMOTE:", Counter(y))

# Inisialisasi SMOTE
smote = SMOTE(random_state=42)

# Resampling data
X_resampled, y_resampled = smote.fit_resample(X, y)

# Cek hasil distribusi setelah SMOTE
print("Distribusi setelah SMOTE:", Counter(y_resampled))
```

```python
df_resampled = pd.concat([pd.DataFrame(X_resampled), pd.DataFrame(y_resampled, columns=["Wasting"])], axis=1)
```

```python
import pandas as pd
df_encoded = pd.get_dummies(df_resampled, columns=['Wasting'])
```

```python
print(df_encoded)
```

```python
df_encoded = df_encoded.rename(columns={
    'Wasting_Normal weight': 'Wasting_Berat Normal',
    'Wasting_Risk of Overweight': 'Wasting_Risiko Kegemukan',
    'Wasting_Severely Underweight': 'Wasting_Sangat Kurus',
    'Wasting_Underweight': 'Wasting_Kurus'
})
print(df_encoded)
```

```python
df_encoded.to_csv('data/data_wasting_smote.csv', index=False)
```
